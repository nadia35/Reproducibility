## Reproducibility Project

This figure (4c) and its data came from this article: https://elifesciences.org/articles/21778#fig7

<b> Import libraries </b>

```r
library(tibble)
library(ggplot2)
library(dplyr)
```

<b> Get the data from the text file </b>

```r
data <- read.table("https://elifesciences.org/download/aHR0cHM6Ly9jZG4uZWxpZmVzY2llbmNlcy5vcmcvYXJ0aWNsZXMvMjE3NzgvZWxpZmUtMjE3NzgtZmlnNC1kYXRhMi12My50eHQ=/elife-21778-fig4-data2-v3.txt?_hash=AGB7%2FDK4Vo9KE53XKG3yUVnNEvXyU%2BsJPfoG8aKDPJY%3D")
```

<b> Convert the table to a tibble </b>

```r
data <- as.tibble(data)
names <- c("top", "fraction", "id")
colnames(data) = names
data <- data[-1,]
```

<b> Adjust the tibble </b>

```r
data$top <- as.numeric(as.character(data$top))
data$fraction <- as.numeric(as.character(data$fraction))
```

<b> Plot the data </b>

```r
ggplot(data, aes(x = top, y = fraction)) + geom_line(aes(color = id)) + theme_classic() + xlab("Top x Ranking Elements") + ylab("COSMIC recall fraction") + scale_color_manual(name = "", labels=c("log-likelihoods", "number of mutations", "conservation"), values = c("logP" = "darkblue", "nMut" = "coral2", "phyloP" = "aquamarine3"))
```
